```python
import requests
from bs4 import BeautifulSoup
```

```python
def process_html_file(file_path):
    # Open and read the content of the HTML file
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Replace &lt; with < and &gt; with >
    content = content.replace('&lt;', '<').replace('&gt;', '>')

    # Return the processed content
    return content
```

```python
# Gets each title of every article from the search page HTML which was manually saved

substring = "/politics/" # restricts to politics articles
politicsArticles = []

for i in range(206):
    path = "CNNHTML/" + str(i) + ".html" # where the html is stored
    html = process_html_file(path)

    soup = BeautifulSoup(html, 'html.parser')
    articles = soup.select('.container__headline-text') # the class for the titles of the articles
    for article in articles:
        article_href = article['data-zjs-href']
        if substring in article_href: # restricts to politcs articles
              politicsArticles.append(article_href)
```

```python
# Saves the file containing the URL of every politics article

file_path = '/content/drive/My Drive/ColabOutput/test.txt'

with open(file_path, 'w') as f:
    for item in politicsArticles:
        f.write(item + '\n')  # Write each item on a new line

print(f"File saved to {file_path}")
```

```python
# Step 2: Define the file path
file_path = 'ColabOutput/test.txt'

# Step 3: Read the file and load it into a list
with open(file_path, 'r') as f:
    politicsArticles = [line.strip() for line in f]  # Use .strip() to remove any extra whitespace or newline characters

file_path_out = 'ColabOutput/allParagraphs.txt'

allParagraphs = []
i = 1
print(len(politicsArticles))

# Open the output file in append mode
for url in politicsArticles:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    text = soup.select('.paragraph')
    
    for paragraph in text:
        with open(file_path_out, 'a') as f:  # Open in append mode ('a') to add new content
            f.write(paragraph.text.strip() + '\n')  # Write each paragraph to the file
    
    print(f"Processed URL {i}: {url}")
    i += 1
```
